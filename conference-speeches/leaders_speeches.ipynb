# Leaders' conference speeches

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
speeches = pd.read_csv('speeches_df.csv')
speeches = speeches[[not x for x in pd.isna(speeches.year)]]
speeches.year = [int(year) for year in speeches.year]
speeches.head()
```

```python
leaders_speeches = [title for title in speeches.title if 'leader' in title.lower().split("'")]
conference = speeches[speeches.title.apply(lambda title: title in leaders_speeches)]
conference.head()
```

```python
party_dict = {'Conservative':(1,'blue'),
              'Labour':(1.5,'red'),
              'Liberal':(2,'gold'),
              'SDP-Liberal Alliance':(2,'gold'),
              'Liberal Democrat':(2,'gold')}
fig,ax = plt.subplots(1, figsize=(8,2))
for row in conference.iterrows():
    ax.scatter(row[1].year, party_dict[row[1].party][0], 
               color=party_dict[row[1].party][1])
ax.set_ylim(0.5,2.5)
plt.show()
```

```python
conference.to_csv('conference.csv')
```
